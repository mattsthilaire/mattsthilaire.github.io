```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
```

```python
def relu(arr):
    return ( (arr > 0) * arr)
```

```python
xs = np.linspace(-10, 10, 10_000)
plt.plot(xs, relu(xs))
```

```python
def dev_relu(arr):
    return (arr > 0) * np.ones(arr.shape)
```

```python
plt.plot(xs, dev_relu(xs))
```
